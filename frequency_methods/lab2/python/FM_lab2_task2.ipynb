```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, a=1, b=1):
    return a * np.exp(-b * t**2)

def f_hat(omega, a=1, b=1):
    return (a / np.sqrt(2 * b)) * np.exp(-omega**2 / (4 * b))

def g(t, c, a=1, b=1):
    return f(t + c, a, b)

def g_hat(omega, c, a=1, b=1):
    return np.exp(1j * omega * c) * f_hat(omega, a, b)

# Параметры
a = 1
b = 1
c_values = [-2, 0, 2]

t = np.linspace(-10, 10, 2000)
omega = np.linspace(-20, 20, 2000)

for c in c_values:
    g_t = g(t, c, a, b)
    G_omega = g_hat(omega, c, a, b)

    # g(t)
    plt.figure(figsize=(12, 4))
    plt.plot(t, g_t)
    plt.title(f"Сдвинутая функция g(t) = f(t + {c})")
    plt.xlabel("t")
    plt.ylabel("g(t)")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"g_function_c{c}.png")
    plt.show()

    # Re ĝ(ω)
    plt.figure(figsize=(12, 4))
    plt.plot(omega, np.real(G_omega), label="Re ĝ(ω)")
    plt.plot(omega, np.imag(G_omega), label="Im ĝ(ω)")
    plt.title(f"Фурье-образ ĝ(ω) при сдвиге c = {c}")
    plt.xlabel("ω")
    plt.ylabel("Значение")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"g_hat_complex_c{c}.png")
    plt.show()

    # |ĝ(ω)|
    plt.figure(figsize=(12, 4))
    plt.plot(omega, np.abs(G_omega))
    plt.title(f"Амплитудный спектр |ĝ(ω)| при сдвиге c = {c}")
    plt.xlabel("ω")
    plt.ylabel("|ĝ(ω)|")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"g_hat_magnitude_c{c}.png")
    plt.show()
```
